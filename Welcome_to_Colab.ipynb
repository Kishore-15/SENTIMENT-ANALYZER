```
import pandas as pd
from textblob import TextBlob

tweets = [
    "I absolutely love this product! ðŸ˜",
    "Terrible service. Never again. ðŸ˜¡",
    "It's okay, nothing special.",
    "Amazing experience! Highly recommend! â­",
    "Worst company ever. Disappointed."
]

def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0.1:
        return 'Positive'
    elif polarity < -0.1:
        return 'Negative'
    return 'Neutral'

# Results
print("SENTIMENT ANALYSIS\n" + "="*50)
for tweet in tweets:
    sentiment = get_sentiment(tweet)

    if sentiment == 'Positive':
        emoji = 'ðŸ˜Š'
    elif sentiment == 'Negative':
        emoji = 'ðŸ˜¡'
    else: # sentiment == 'Neutral'
        emoji = 'ðŸ˜'

    print(f"\nTweet: {tweet}")
    print(f"Sentiment: {sentiment} {emoji}")

print("\n" + "="*50)
print("âœ“ Done!")
```
